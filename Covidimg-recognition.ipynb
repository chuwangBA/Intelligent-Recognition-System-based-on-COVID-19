## 3. Build Baseline Models

```python
#baseline Models check accuracy
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score,accuracy_score
from sklearn.model_selection import cross_val_score
#define function
def baseline_models_results(X_train,y_train,X_test):
    lr=LogisticRegression()
    knn=KNeighborsClassifier()
    svc=SVC()
    dt=DecisionTreeClassifier()
    rf=RandomForestClassifier()
    gbdt=GradientBoostingClassifier()
    for i in [lr,knn,svc,dt,rf,gbdt]:
        i.fit(X_train,y_train)
        y_predict=i.predict(X_test)
        print('{}在训练集上表现为：{}，在测试集上表现为：{}'.format(i,roc_auc_score(y_train,i.predict(X_train)),roc_auc_score(y_test,y_predict)))
```

## 4. Feature Engineering

```python
#对给定的数据集进行GBDT特征组合
def gbdt_fc(X_train,y_train,X_test,features,depth):
    gbdt_clf = GradientBoostingClassifier(n_estimators=features,
                                      max_depth=depth,
                                      random_state=0)
    gbdt_clf.fit(X_train,y_train)
    # 注意这里使用的是apply函数
    feature_extract_train_clf = gbdt_clf.apply(X_train)#对原数据集进行特征组合
    feature_extract_test_clf = gbdt_clf.apply(X_test)

    feature_extract_train_clf = pd.DataFrame(feature_extract_train_clf.squeeze(2))
    feature_extract_test_clf = pd.DataFrame(feature_extract_test_clf.squeeze(2))
    #先合并再编码
    feature_combined=pd.concat([feature_extract_train_clf,feature_extract_test_clf],axis=0)
    feature_combined_d=pd.get_dummies(feature_combined.astype('category'),drop_first=True)
    #拆分开
    return feature_combined_d[:559],feature_combined_d[559:]
```

---
### FE Method1：qcut 500 features->get_dummies->generate 4000 features->Wrapper FS->Low Performance
- Limited samples,embedded method may not show good result in too many features
- Wrapper/Filter Feature Selection will avoid above problems
- Results:Lower Performance than Baseline models---Out

```python
#Concat
X_combine=pd.concat([X_train,X_test],axis=0)
# X_train=X_combine[:559]
# X_test=X_combine[559:]

#Qcut for concated data
X_combine_box=pd.DataFrame()
for i in X_combine.columns:
    X_combine_box[i]=pd.qcut(X_combine.iloc[:,int(i)],labels=False,q=10,duplicates='drop')#等频分箱，等距就是将pd.qcut(q=?)改为pd.cut(bins=?)

#Trandform to category for get_dummies
X_combine_box=X_combine_box.astype('category')

#get_dummies
X_combine_box_d=pd.get_dummies(X_combine_box,drop_first=True)

#Split to original train test set
X_train_d=X_combine_box_d[:559]
X_test_d=X_combine_box_d[559:]
```

```python
#Wrapper-Feature Selection---GBDT
gbdt=GradientBoostingClassifier()
from sklearn.feature_selection import RFECV
select=RFECV(gbdt,step=200,cv=3,scoring='roc_auc',n_jobs=-1)
select.fit(X_train_d,y_train)
print('对于GBDT使用包裹法选择出最优特征数：{}个'.format(select.n_features_))
print('最优特征集的cv-score：{}'.format(select.grid_scores_.max()))
plt.plot(range(1,1+len(select.grid_scores_)),select.grid_scores_)
plt.xlabel('Number of Subset Selected')
plt.ylabel('cv-score for each subsets')
plt.title('GBDT-Recursive Feature Elimination')
plt.show()
```

```python
#Wrapper-Feature Selection---LogisticRegression
lr=LogisticRegression()
select1=RFECV(lr,step=200,cv=3,scoring='roc_auc',n_jobs=-1)
select1.fit(X_train_d,y_train)
print('对于LogisticRegression使用包裹法选择出最优特征数：{}个'.format(select1.n_features_))
print('最优特征集的cv-score：{}'.format(select1.grid_scores_.max()))
plt.plot(range(1,1+len(select1.grid_scores_)),select1.grid_scores_)
plt.xlabel('Number of Subset Selected')
plt.ylabel('cv-score for each subsets')
plt.title('LogisticRegression-Recursive Feature Elimination')
plt.show()
```

```python
#GBDT Feature Crosses
X_train_fe2,X_test_fe2=gbdt_fc(select2.transform(X_train),y_train.values,select2.transform(X_test),features=50,depth=2)
```

```python
#PCA dimension reduction
from sklearn.decomposition import KernelPCA
pca=KernelPCA(kernel='rbf')
pca.fit(X_train_fe2,y_train)
plt.plot(((pca.lambdas_)/(pca.lambdas_.sum())).cumsum())
plt.show()
# make threshold=0.85
for i in range(len(pca.lambdas_)):
    if pca.lambdas_[:i+1].sum()/(pca.lambdas_.sum())>=0.85:
        n_components=i
        break
    else:
        continue
print('n_components equal to:',n_components)
```

```python
#PCA dimension reduction--n_components=50
pca_fe2=KernelPCA(kernel='rbf',n_components=n_components)
X_train_fe2_pca=pca_fe2.fit_transform(X_train_fe2)
X_test_fe2_pca=pca_fe2.transform(X_test_fe2)
```

```python
##Plot learning curve
from sklearn.model_selection import learning_curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()
 
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")
 
    plt.legend(loc="best")
    return plt
```

## 6. Adjusting threshold of selected model

```python
#return the predicted label according to different threshold
def recall_threshold(y_proba,threshold):
    predictions=[]
    for i in y_proba:
        if i<threshold:
            predictions.append(0)
        else:
            predictions.append(1)
    return np.array(predictions)
#return target recall_score acc_score lists according to the specified thresholds
def recall_accuracy_lst(y_true,y_pred_proba,threshold):#y_true and y_pred_proba are both nd.array,threshold is a list 0-1
    recall_lst=[]
    acc_lst=[]
    for i in threshold:
        recall_lst.append(recall_score(y_true,recall_threshold(y_pred_proba,i)))
    for j in threshold:
        acc_lst.append(accuracy_score(y_true,recall_threshold(y_pred_proba,j)))
    return recall_lst,acc_lst
```

```python
#Plot train set recall_score and acc_score of five models by changing the threshold
fig=plt.figure(figsize=(20,10))
n=1
for i,j in zip([y_tr_pred_gbdt,y_tr_pred_rf,y_tr_pred_lr,y_tr_pred_svc,y_tr_pred_knn],['GBDT','RF','LR','SVC','KNN']):
    train_recall,train_acc=recall_accuracy_lst(y_train.values,i,np.arange(0,1,0.01))
    ax=fig.add_subplot(2,3,n)
    n=n+1
    ax.plot(train_recall,label='recall')
    ax.plot(train_acc,label='accuracy')
    ax.set_xlabel('number of thresholds')
    ax.set_ylabel('score')
    ax.set_title(j+' Train Set Recall and Accuracy')
```
